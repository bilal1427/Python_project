```python
# Analyze how temperature change has varied globally over time (1961–2022).

import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/sample_data/Surface_temp_csv.csv'
extended_unique_data = pd.read_csv(file_path)

# Check for valid year columns
years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in years if year in extended_unique_data.columns]

global_trend = extended_unique_data[valid_years].mean()

plt.figure(figsize=(12, 6))
plt.plot(valid_years, global_trend, marker='o', color='blue')
plt.title('Global Temperature Change Trend (1961–2022)')
plt.xlabel('Year')
plt.ylabel('Temperature Change (°C)')
plt.xticks(ticks=range(0, len(valid_years), 5), labels=valid_years[::5], rotation=0)

plt.grid(True)
plt.show()
```

```python
#Compare temperature changes across different regions (Sub-Region).

regional_avg = extended_unique_data.groupby('Sub-Region')['2022'].mean()

regional_avg.sort_values().plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Average Temperature Change by Sub-Region (2022)')
plt.xlabel('Sub-Region')
plt.ylabel('Temperature Change (°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Investigate temperature trends across income groups.

years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in years if year in extended_unique_data.columns]

# Compute average temperature change by income group
income_group_avg = extended_unique_data.groupby('Income Group')[valid_years].mean()

# Transpose the data for plotting
income_group_avg = income_group_avg.T

plt.figure(figsize=(12, 6))
plt.stackplot(income_group_avg.index, income_group_avg.values.T, labels=income_group_avg.columns, alpha=0.8)

plt.title('Temperature Trends by Income Group (1961–2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Change (°C)', fontsize=12)
plt.legend(title='Income Group', loc='upper left')
plt.grid(True, alpha=0.5)
plt.xticks(ticks=range(0, len(valid_years), 5), labels=valid_years[::5], rotation=0)
plt.show()

```

```python
#Identify countries with extreme temperature change values.
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in years if year in data.columns]

data['Average_Temperature_Change'] = data[valid_years].mean(axis=1)

# Get top 10 and bottom 10 countries
top_countries = data.nlargest(10, 'Average_Temperature_Change')
bottom_countries = data.nsmallest(10, 'Average_Temperature_Change')

# Combine top and bottom countries
outliers = pd.concat([top_countries, bottom_countries])

# Plot bar graph
plt.figure(figsize=(12, 6))
plt.bar(outliers['Country'], outliers['Average_Temperature_Change'], color='skyblue')
plt.title('Top and Bottom 10 Countries by Average Temperature Change (1961–2022)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Temperature Change (°C)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Explore the correlation between temperature changes and population density (Density (KM²))

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

# Check for valid year columns
years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in data.columns if year in years]

# Calculate average temperature change
data['Average_Temperature_Change'] = data[valid_years].mean(axis=1)

# Select all numerical columns for correlation analysis
numerical_columns = ['Density (KM²)', 'Area (KM²)', 'Average_Temperature_Change']
correlation_matrix = data[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=1, square=True)
plt.title('Correlation Heatmap of Numerical Features', fontsize=14)
plt.show()


```

```python
#Determine countries with the highest/lowest temperature changes.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

# Check for valid year columns
years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in data.columns if year in years]

# Calculate average temperature change
data['Average_Temperature_Change'] = data[valid_years].mean(axis=1)

# Get top 10 and bottom 10 countries
top_countries = data.nlargest(10, 'Average_Temperature_Change')
bottom_countries = data.nsmallest(10, 'Average_Temperature_Change')

# Plot horizontal bar charts for both
plt.figure(figsize=(14, 6))

# Top 10 countries
plt.subplot(1, 2, 1)
plt.barh(top_countries['Country'], top_countries['Average_Temperature_Change'], color='red')
plt.title('Top 10 Countries by Avg Temperature Change')
plt.xlabel('Average Temperature Change (°C)')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Bottom 10 countries
plt.subplot(1, 2, 2)
plt.barh(bottom_countries['Country'], bottom_countries['Average_Temperature_Change'], color='blue')
plt.title('Bottom 10 Countries by Avg Temperature Change')
plt.xlabel('Average Temperature Change (°C)')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

```

```python
#Analyze the relationship between country size and temperature changes.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

# Check for valid year columns
years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in data.columns if year in years]

# Calculate average temperature change
data['Average_Temperature_Change'] = data[valid_years].mean(axis=1)

# Define bins for country size (Area in KM²)
bins = [0, 10000, 100000, 1000000, 10000000, data['Area (KM²)'].max()]
labels = ['<10K', '10K-100K', '100K-1M', '1M-10M', '>10M']

# Bin the data
data['Area_Bin'] = pd.cut(data['Area (KM²)'], bins=bins, labels=labels, include_lowest=True)

# Calculate average temperature change for each bin
binned_avg_temp_change = data.groupby('Area_Bin')['Average_Temperature_Change'].mean()

# Plot the binned bar chart
plt.figure(figsize=(10, 6))
binned_avg_temp_change.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Temperature Change by Country Size ', fontsize=16)
plt.xlabel('Country Size (KM²)', fontsize=12)
plt.ylabel('Average Temperature Change (°C)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()

```

```python
#Identify years with significant temperature changes.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

# Check for valid year columns
years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in data.columns if year in years]

# Calculate global average temperature change for each year
global_avg_temp = data[valid_years].mean()

# Calculate year-on-year differences
year_diff = global_avg_temp.diff()

# Identify years with the highest increase and decrease
max_increase_year = year_diff.idxmax()
max_decrease_year = year_diff.idxmin()

# Plot the year-on-year changes
plt.figure(figsize=(14, 8))
plt.plot(years[1:], year_diff[1:], label='Year-on-Year Change', color='blue', marker='o')
plt.axvline(max_increase_year, color='green', linestyle='--', label=f'Max Increase: {max_increase_year}')
plt.axvline(max_decrease_year, color='red', linestyle='--', label=f'Max Decrease: {max_decrease_year}')
plt.title('Year-on-Year Variability in Global Average Temperature (1961–2022)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Change (°C)', fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.xticks(rotation=45)
plt.show()


```

```python
#A Proportion of Countries in Sub-Regions

import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/sample_data/Surface_temp_csv.csv'
data = pd.read_csv(file_path)

years = [str(year) for year in range(1961, 2023)]
valid_years = [year for year in data.columns if year in years]

data['Average_Temperature_Change'] = data[valid_years].mean(axis=1)
sub_region_avg_temp = data.groupby('Sub-Region')['Average_Temperature_Change'].mean()

sub_region_avg_temp = sub_region_avg_temp[sub_region_avg_temp > 0]  # Keep only positive averages

# Plot the pie chart
plt.figure(figsize=(10, 8))
sub_region_avg_temp.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='viridis', legend=False)
plt.title('Average Temperature Change by Sub-Region (1961–2022)', fontsize=14)
plt.ylabel('')  # Remove default ylabel
plt.show()



```

```python
import seaborn as sns

# Filter data for Sub-Saharan Africa
sub_saharan_africa = data[data['Sub-Region'] == 'Sub-Saharan Africa']

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(sub_saharan_africa['Density (KM²)'], bins=20, kde=True, color='purple')
plt.title('Population Density Distribution in Sub-Saharan Africa', fontsize=14)
plt.xlabel('Population Density (per KM²)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
